```python
from gensim.models import Word2Vec

corpus = [
    ["I", "love", "machine", "learning"],
    ["I", "love", "Python"],
    ["machine", "learning", "is", "great"]
]

# Train model
model = Word2Vec(
    sentences=corpus,
    vector_size=100,   # embedding dimensions
    window=5,          # context window size
    min_count=1,       # ignore words with freq < 1
    sg=0               # 0 = CBOW, 1 = Skip-Gram
)

# Get vector for a word
vector = model.wv["machine"]
print(vector.shape)   # (100,)

# Find similar words
model.wv.most_similar("machine")
```

| Parameter     | Description                                                                 |
|--------------|-----------------------------------------------------------------------------|
| vector_size  | Dimension of word vectors (typically 100â€“300)                              |
| window       | Context window size (number of words around the target word)              |
| min_count    | Ignore words with frequency lower than this value                         |
| sg           | 0 = CBOW, 1 = Skip-Gram                                                    |
| epochs       | Number of training iterations over the corpus                             |
| workers      | Number of parallel threads used during training                           |
